```python
people = {
    "first": ["Corey", "Jane", "John"],
    "last": ["Schafer", "Doe", "Doe"],
    "email": ["CoreyMSchafer@gmail.com", "JaneDoe@email.com", "JohnDoe@email.com"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
filt = (df['last'] == 'Schafer')
```

```python
df[filt]
```

```python
df.loc[filt, 'email']
```

```python
file = (df['last'] == 'Doe') & (df['first'] == 'John')
```

```python
df[file]
```

```python
df.loc[~file, 'email']
```
