```python
people = {
    "first": ['Corey', 'Jane', 'John', 'Adam'],
    "last": ['Schafer', 'Doe', 'Doe', 'Doe'],
    "email": ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', 'AdamDoe@gmail.com']
}

import pandas as pd

df = pd.DataFrame(people)

df
```

```python
df.sort_values(by='last')
```

```python
df.sort_values(by='last', ascending = False)
```

```python
df.sort_values(by=['last', 'first'], ascending = False) #last first and first it last equal
```

```python
df.sort_values(by=['last', 'first'], ascending = [False,True], inplace = True) #corresponds to columns in 'by'
```
